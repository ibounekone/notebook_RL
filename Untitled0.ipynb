```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#lien de la donnée
dataset_url = "https://raw.githubusercontent.com/OpenClassrooms-Student-Center/8063076-Initiez-vous-au-Machine-Learning/master/data/advertising.csv"
#dataframe de lecture de la donnée
df = pd.read_csv(dataset_url)
#initialisation des variable
X=df[['tv','radio','journaux']]
y=df.ventes
#visualisation du nuage de points

fig, ax =plt.subplots(1,3,figsize = (15,5))

#affichage du nuage de points de tv
plt.subplot(1,3,1)
sns.regplot(x=df[['tv']], y=y)
plt.xlabel('TV')
plt.ylabel('ventes')
plt.title('ventes= a * tv + b')
plt.grid()
sns.despine()

#affichage du nuage de pounts radio
plt.subplot(1,3,2)
sns.regplot(x=df[['radio']],y=y)
plt.xlabel('RADIO')
plt.ylabel('Ventes')
plt.title('ventes= a * radio + b')
plt.grid()
sns.despine()

#affichage du nuage de points journaux
plt.subplot(1,3,3)
sns.regplot(x=df[['journaux']],y=y)
plt.xlabel('JOURNAUX')
plt.ylabel('Ventes')
plt.title('vente= a * journaux + b')
plt.grid()
sns.despine()

plt.tight_layout()
plt.show()
#affichage de 5 premiers éléments
#df.head()
#df.describe()
#df.info()
df.corr()

#choix du modèle
from sklearn.linear_model import LinearRegression
#instanciation
reg=LinearRegression()

#division du jeu de données
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.20, random_state=42)

#entrainement du modèle
reg.fit(X_train, y_train)

#prediction avec la donnée test
y_pred_test=reg.predict(X_test)

#calcule de l'écart entre les valeurs prédite et les valeurs réelles
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

print(f"RMSE: {mean_squared_error(y_test, y_pred_test)}")
print(f"MAPE: {mean_absolute_percentage_error(y_test, y_pred_test)}")

#ajout d'une puissance de préditeur
df['tv2']=df.tv**2

#normalisation de l'amplitude pour standartiser les coéficients à cause de la quadrature
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df)
data_array=scaler.transform(df)
#réecreture du tableau data_array en dataframe
df=pd.DataFrame(data_array,columns=['tv','radio','journaux','ventes','tv2'])
df.describe().loc[['min','max']]
X=df[['tv','radio','journaux','tv2']]
y=df.ventes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
reg.fit(X_train, y_train)
y_hat_test=reg.predict(X_test)

print("**********")
print(f"Coéfficient: {reg.coef_}")
print(f"RMSE: {mean_squared_error(y_test, y_hat_test)}")
print(f"MAPE: {mean_absolute_percentage_error(y_test, y_hat_test)}")

#ajout d'un cumul de predicteur tv * radio
df['tv_radio']=df.tv*df.radio
scaler=MinMaxScaler()
scaler.fit(df)
data_array=scaler.transform(df)
#réecreture du tableau data_array en dataframe
df=pd.DataFrame(data_array,columns=['tv','radio','journaux','ventes','tv2','tv_radio'])
df.describe().loc[['min','max']]
X=df[['tv','radio','journaux','tv2','tv_radio']]
y=df.ventes
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
reg.fit(X_train, y_train)
y_hat_2_test=reg.predict(X_test)

print("**********")
print(f"Coéfficient: {reg.coef_}")
print(f"RMSE: {mean_squared_error(y_test, y_hat_2_test)}")
print(f"MAPE: {mean_absolute_percentage_error(y_test, y_hat_2_test)}")


#visualisation du nuage de points de tv2 et tv_radio

#fig, ax =plt.subplots(1,2,figsize = (15,5))

#affichage du nuage de points de tv
#plt.subplot(1,2,1)
#sns.regplot(x=df[['tv2']], y=y)
#plt.xlabel('TV2')
#plt.ylabel('ventes')
#plt.title('ventes= a * tv2 + b')
#plt.grid()
#sns.despine()

#affichage du nuage de pounts radio
#plt.subplot(1,2,2)
#sns.regplot(x=df[['tv_radio']],y=y)
#plt.xlabel('TV_RADIO')
#plt.ylabel('Ventes')
#plt.title('ventes= a * tv * radio + b')
#plt.grid()
#sns.despine()

#plt.tight_layout()
#plt.show()
```
